```python
# importing sentence transformer
from sentence_transformers import SentenceTransformer
```

```python
# import pandas for df operation
import pandas as pd
# import json for load .json file 
import json
```

```python
# importing torch for tensor operations
import torch
```

```python
# open appointments file in read
with open('../appointments_data.json', 'r') as f:
    # load json file into data
    data = json.load(f)
```

```python
# dowload pretrained qwen embeding with intialize embed mode
model = SentenceTransformer("Qwen/Qwen3-Embedding-0.6B")
```

```python
# get keys list using dict comprehession
keys = [ key for key in data.keys() ]
```

```python
# encode keys to get embedding
keys_embedding = model.encode(keys)
```

```python
len(keys_embedding)
```

```python
# print embedding
for key_embed in keys_embedding:
    print(key_embed)
```

```python
# query initialize
query = "Who is doctor of Brooke"
```

```python
# get embeding of query
query_embedding = model.encode(query)
```

```python
query_embedding
```

```python
# find similarity between query and keys
similarity = model.similarity(query_embedding, keys_embedding)
```

```python
len(similarity[0])
```

```python
similarity = torch.tensor(similarity)
```

```python
# topk gives top score and index
scores, top10_indices = torch.topk(similarity, k=10)
```

```python
top10_indices
```

```python
scores
```

```python
# intialize dict keys 10 for top 10
keys_10 = {}
# use list comprehession to get top ten keys
top10_key =  [keys[i] for  i in top10_indices[0]]
# use list comprehession to get top ten score
top10_scores = [i for i in scores[0]]

```

```python
top10_scores
```

```python
top10_key
```

```python
# here iterate upto len of top10 keys
for i in range(len(top10_key)):
    # update dict with key and value as key and score
    keys_10[top10_key[i]] = top10_scores[i]
```
